```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(data_set.isnull())
plt.show()

# white => having null value
# black => value are not missing 
```

```python
# We will check that how much data is missing in percentage 
null_val = data_set.isnull().sum()/data_set.shape[0]*100
null_val
```

```python
# if data percentage is more than 20% than delete it
drop_colns = null_val[null_val>17].keys()
print(drop_colns)  # these columns are creating blunder -> having lots of missing values

# There is another column "LotFrontage" missing data is 17.73972602739726 
# for now we will remove that column from the data set
```

```python
update_data_set = data_set.drop(columns= drop_colns)
# Index(['LotFrontage', 'Alley', 'MasVnrType', 'FireplaceQu', 'PoolQC', 'Fence',
#        'MiscFeature'],   dtype='object')  These columns are droped or removed from teh dataset 
update_data_set.shape
update_data_set.head()
```

```python
# Checking the missing values from update_data_set
plt.figure(figsize=(20,20))
sns.heatmap(update_data_set.isnull())
plt.show()
# Here we have removed the impurity
```

```python
# to delete row of missing values  use  <dataframe>.dropna() function of pandas
new_data_clean = update_data_set.dropna()  # by default row  
print(new_data_clean.shape)  # this shape is for new_data_clean data set
print(update_data_set.shape)  # this shape is for update_data_set data set
```

```python
# new cleaned data 
plt.figure(figsize=(20,20))
sns.heatmap(new_data_clean.isnull())
plt.show()
```

```python
numerical_variables = ['MSSubClass', 'LotArea', 'OverallQual', 'OverallCond',
       'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2',
       'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
       'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
       'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
       'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF',
       'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal',
       'MoSold', 'YrSold', 'SalePrice']

plt.figure(figsize=(25,25))
for i, var in enumerate(numerical_variables):  # var is the numerical_variables  and i is iteration / increment
    plt.subplot(9,4,i+1)                        # 9 rows and 4 columns  , i=> is for index starts from 1
    sns.displot(data_set[var],bins = 20 , rug = True)
    sns.displot(new_data_clean[var], bins = 20, rug = True)
```

```python
# for categorical data 
categorical = new_data_clean.select_dtypes(include=["object"]).columns
```

```python
# Here count the category of MSZoning => data_set['MSZoning'].value_counts()  output is like
# MSZoning
# RL         1151
# RM          218
# FV           65
# RH           16
# C (all)      10

# new_data_clean.shape[0]  rows 
# data_set['MSZoning'].value_counts()/new_data_clean.shape[0]*100  percentage 
new = pd.concat([data_set['MSZoning'].value_counts()/data_set.shape[0]*100,
          new_data_clean['MSZoning'].value_counts()/new_data_clean.shape[0]*100],axis = 1,
          keys=["MSZoning_org","MSZoning_clean"])    # creating the data frame 
print(type(new))
new

# use to see the difference between between cleaned and original data set 
```

```python
def Categorical_data(var):
    return(pd.concat([data_set[var].value_counts()/data_set.shape[0]*100,
          new_data_clean[var].value_counts()/new_data_clean.shape[0]*100],axis = 1,
          keys=[var+"_org",var+"_clean"]))   # creating the data frame 
```

```python
for i in categorical:
    print(Categorical_data(i))
    print("\n")
```
